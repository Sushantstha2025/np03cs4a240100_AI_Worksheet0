```python
def sum_nested_list(nested_list):
    '''
    it returns the sum of the nested list regardless of the depth of the list. It uses recursion to continuously call itself to unpack the list and use its element to add to the total. To check if the element is a list or just a value, we use isinstance() to determine it.

    param: list that is nested

    returns: the sum of all the value of the nested list regardless of its depth.
    '''
    total = 0
    for element in nested_list:
        if isinstance(element, list):   # takes two params, value, and what we are comparing it against. In this case, we are comparing if it is a list or not.
            total += sum_nested_list(element)
        else:
            total += element

    return total

print(sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]]))
```

```python
def generate_permutations(s):
    result = []
    if len(s) == 1:
        return [s]

    else:
        for i, char in enumerate(s):
            remaining = s[:i] + s[i+1: ]

            for perm in generate_permutations(remaining):
                result.append(char + perm)


        return result

string = "abc"
all_perms = generate_permutations(string)
print(all_perms)
```

```python
def calculate_directory_size(directory):
    total = 0
    for item, value in directory.items():
        if isinstance(value, dict):
            total += calculate_directory_size(value)

        else:
            total += value

    return total


directory_structure = {
        "file1.txt": 200,
        "file2.txt": 300,
        "subdir1": {
            "file3.txt": 400,
            "file4.txt": 100
        },
        "subdir2": {
            "subsubdir1": {
                "file5.txt": 250
            },
            "file6.txt": 150
        }
    }
print(calculate_directory_size(directory_structure))
```

```python
# 1. Write a function min_coins(coins, amount) that:
# • Uses DP to calculate the minimum number of coins needed to make up the
# amount.
# 2. Test with coins = [1, 2, 5] and amount = 11. The result should be 3 (using coins
# [5, 5, 1]).

def min_coins(coins, amount):
    inf = float('inf')
    dp = [inf] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != inf else -1


print(min_coins([1, 2, 5], 11))
```

```python
# Scenario: Given two strings, find the length of their longest common subsequence (LCS).
# This is useful in text comparison.
# Task:
# 1. Write a function longest_common_subsequence(s1, s2) that:
# • Uses DP to find the length of the LCS of two strings s1 and s2.
# 2. Test with strings like "abcde" and "ace"; the LCS length should be 3 ("ace").


def longest_common_subsequence(s1, s2):
    n, m = len(s1), len(s2)

    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]

print(longest_common_subsequence("abcdef", "acef"))
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


print(knapsack([1, 3, 4, 5],[1, 4, 5, 7],7))
```
